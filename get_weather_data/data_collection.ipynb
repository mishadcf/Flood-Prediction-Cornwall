```python


def get_weather_data(station_id, start_date, end_date):
    url = "https://meteostat.p.rapidapi.com/stations/hourly"
    querystring = {
        "station": station_id,
        "start": start_date,
        "end": end_date,
        "tz": "Europe/Berlin"
    }
    headers = {
        "X-RapidAPI-Key": "f1b37d13b4msh85c650be49c6e89p10a13ejsnd5c48470b3da",
        "X-RapidAPI-Host": "meteostat.p.rapidapi.com"
    }
    
    response = requests.get(url, headers=headers, params=querystring)
    return response.json()

stations = ["EGUH0", "03772"]
start_date = "2013-04-30"
end_date = "2023-08-31"

for station in stations:
    data = get_weather_data(station, start_date, end_date)
    
    # Convert to pandas DataFrame
    df = pd.DataFrame(data.get('data', []))
    
    # Display first few rows
    print(f"Weather data for station {station}:")
    print(df.head())
    
    # Save to CSV
    filename = f"weather_data_{station}.csv"
    df.to_csv(filename, index=False)
    print(f"Saved data to {filename}")
    print("-----------------------------------------------------")
```
