```python
river_info = get_top_level_info_river_gauges()
```

```python
river_info
```

```python

river_coords = {}
for _, data in river_info.iterrows():
    name = data['name']
    coordinates = (data['longitude'], data['latitude'])
    river_coords[name] = coordinates
```

```python
river_coords
```

```python
import folium

# Create a map centered around Cornwall
map = folium.Map(location=[50.2660, -5.0527], zoom_start=10)
```

```python

# Create a map centered around an example location
map = folium.Map(location=[50.2660, -5.0527], zoom_start=10)

# Add markers for river gauges using CircleMarker for simplicity
for name, (lat, lon) in river_coords.items():
    folium.CircleMarker(
        location=[lat, lon],
        radius=5,  # size of the circle markers
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.7,
        popup=name
    ).add_to(map)

# Add markers for weather stations using CircleMarker with a different color
for name, (lat, lon) in weather_station_coords.items():
    folium.CircleMarker(
        location=[lat, lon],
        radius=5,
        color='green',
        fill=True,
        fill_color='green',
        fill_opacity=0.7,
        popup=name
    ).add_to(map)

# Save and display the map
map.save('index.html')
map
```

```python
import geopandas as gpd
from shapely.geometry import Point
import contextily as cx
import matplotlib.pyplot as plt

# River gauge data
river_gauge_data = river_coords

# Weather station data
weather_stations_data = weather_station_coords

# Convert to DataFrames
river_gauges_df = pd.DataFrame(river_gauge_data).T
river_gauges_df.columns = ['Longitude', 'Latitude']

weather_stations_df = pd.DataFrame(weather_stations_data).T
weather_stations_df.columns = ['Longitude', 'Latitude']

# Convert to GeoDataFrames
river_gauges_geometry = [Point(xy) for xy in zip(river_gauges_df['Longitude'], river_gauges_df['Latitude'])]
weather_stations_geometry = [Point(xy) for xy in zip(weather_stations_df['Longitude'], weather_stations_df['Latitude'])]

river_gauges_gdf = gpd.GeoDataFrame(river_gauges_df, geometry=river_gauges_geometry, crs="EPSG:4326")
weather_stations_gdf = gpd.GeoDataFrame(weather_stations_df, geometry=weather_stations_geometry, crs="EPSG:4326")

# Convert to web mercator for basemap compatibility
river_gauges_gdf = river_gauges_gdf.to_crs(epsg=3857)
weather_stations_gdf = weather_stations_gdf.to_crs(epsg=3857)

# Plotting
fig, ax = plt.subplots(figsize=(10, 10))

# Plot river gauges and weather stations
river_gauges_gdf.plot(ax=ax, color='blue', label='River Gauges', markersize=50)
weather_stations_gdf.plot(ax=ax, color='red', label='Weather Stations', markersize=50)

# Add basemap
cx.add_basemap(ax, source=cx.providers.Stamen.TerrainBackground)

# Add title and legend
ax.set_title('River Gauges and Weather Stations in Cornwall, UK', fontsize=16)
plt.legend()

plt.show()
```
